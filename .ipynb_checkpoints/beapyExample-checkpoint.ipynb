## Methods for searching for data

* `getDataSetList()`: returns the available datasets.
* `getParameterList(dataSetName)`: returns the parameters of the specified dataset.
* `getParameterValues(dataSetName,ParameterName)`: returns the values accepted for a parameter of the specified dataset.

#### Example: Relative per capita incomes in the US

```python
# Get per capita personal income at the state level for all years.
frame = bea.getRegionalData(KeyCode='PCPI_SI',GeoFips = 'STATE', Year = 'ALL')

# For each state including Washington, D.C., find the percentage difference between state pc income and US pc income.
for state in frame.columns:
    f = 100*(frame[state] - frame['United States'])/frame['United States']
    f.plot(grid=True)
```

```python
import datetime
import requests
import numpy as np
import pandas as pd
import sys

def convertDate(dateString,Frequency):

    '''Function for converting the date strings from BEA with quarter indicators into datetime format'''

    if Frequency=='A':
        month='01'
    elif Frequency=='Q':
        if dateString[-1]=='1':
            month='01'
        elif dateString[-1]=='2':
            month='04'
        elif dateString[-1]=='3':
            month='07'
        else:
            month='10'
    return datetime.datetime.strptime(dateString[0:4]+'-'+month+'-01','%Y-%m-%d')
```
